```python
#inspecting the dataset
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('../data/heart.csv')  # Adjust path if needed

# 1. DATASET SHAPE
print("=" * 50)
print("DATASET SHAPE")
print("=" * 50)
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
print(f"Total cells: {df.shape[0] * df.shape[1]}")

# 2. COLUMN NAMES & DATA TYPES
print("\n" + "=" * 50)
print("COLUMN NAMES & DATA TYPES")
print("=" * 50)
print(df.dtypes)

# 3. FIRST & LAST 5 ROWS
print("\n" + "=" * 50)
print("FIRST 5 ROWS")
print("=" * 50)
print(df.head())

print("\n" + "=" * 50)
print("LAST 5 ROWS")
print("=" * 50)
print(df.tail())

# 4. BASIC STATISTICS
print("\n" + "=" * 50)
print("BASIC STATISTICS (Numerical Columns)")
print("=" * 50)
print(df.describe())

# 5. MISSING VALUES
print("\n" + "=" * 50)
print("MISSING VALUES")
print("=" * 50)
print(df.isnull().sum())
print(f"\nTotal missing values: {df.isnull().sum().sum()}")

# 6. DATA INFO (Memory usage, column info)
print("\n" + "=" * 50)
print("DATA INFO")
print("=" * 50)
df.info()

# 7. UNIQUE VALUES PER COLUMN
print("\n" + "=" * 50)
print("UNIQUE VALUES PER COLUMN")
print("=" * 50)
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")

# 8. VALUE COUNTS FOR CATEGORICAL COLUMNS
print("\n" + "=" * 50)
print("VALUE COUNTS (Categorical Columns)")
print("=" * 50)
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\n{col}:")
    print(df[col].value_counts())
```
